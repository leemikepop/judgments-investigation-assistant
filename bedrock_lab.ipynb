```python
investigation_subject = "台灣世曦工程顧問"
jud_content = """
臺灣士林地方法院民事判決　　　　　　　 103年度建字第65號
原　　　告　陳信樟建築師事務所即陳信樟
訴訟代理人　楊理安律師
　　　　　　林信和律師
複 代 理人　胡閏祺律師
　　　　　　謝宗哲律師
被　　　告　國防部軍備局工程營產中心
法定代理人　吳孟武
訴訟代理人　張仁興律師
　　　　　　張倍齊律師
複 代 理人　龔君安
受 告 知人　中華工程股份有限公司
法定代理人　沈慶京
受 告 知人　浩漢忠孝工程股份有限公司
法定代理人　林正雄
上列當事人間請求給付規劃設計費事件，本院於民國105 年3月1
7日言詞辯論終結，判決如下：
    主  文
被告應給付原告新臺幣伍佰陸拾捌萬肆仟肆佰伍拾伍元，及自民
國一○三年八月十二日起至清償日止，按年息百分之五計算之利
息。
訴訟費用由被告負擔。
    事實及理由
壹、程序方面：
一、按當事人得於訴訟繫屬中，將訴訟告知於因自己敗訴而有法
    律上利害關係之第三人，民事訴訟法第65條第1 項定有明文
    。經查，被告於本件訴訟繫屬中之民國103 年9 月16日具狀
    主張訴外人中華工程股份有限公司（下稱中華工程公司）、
    浩漢忠孝工程股份有限公司（下稱浩漢忠孝公司）與原告為
    統包團隊（下稱統包團隊），因原告對被告提起本件請求給
    付規劃設計費之訴訟結果，會影響統包團隊就剩餘未受領之
    工程尾款分配，有法律上之利害關係為由，聲請本院向中華
    工程公司及浩漢忠孝公司為訴訟之告知，上開告知訴訟之書
    狀業於103 年10月16日送達受告知人，合先說明。
二、按訴狀送達後，原告不得將原訴變更或追加他訴。但請求之
    基礎事實同一者、擴張或減縮應受判決事項之聲明者，不在
    此限，民事訴訟法第255 條第1 項第2 款、第3 款分別定有
    明文。本件原告起訴時原聲明為：被告應給付原告新臺幣（
    下同）570 萬868 元，及自起訴狀繕本送達翌日起至清償日
    止，按年息百分之五計算之利息（下稱法定遲延利息）。嗣
    於103 年10月16日具狀將聲明變更為：被告應給付原告568
    萬4,455 元，及自起訴狀繕本送達翌日起至清償日止計算之
    法定遲延利息。（見本院卷一第83頁）。核原告前開所為，
    均係基於請求被告履行兩造間簽訂之工程採購契約並給付規
    劃設計費尾款之同一基礎事實，僅係減縮請求被告給付之數
    額，揆諸首揭規定，在程序上應予准許。
貳、實體方面：
一、原告主張：伊與中華工程公司、浩漢忠孝公司共同投標被告
    之「國醫中心職務官舍新建工程（第二階段）統包工程」（
    下稱系爭統包工程），並於93年8 月17日共同與被告簽訂「
    國防部軍備局工程營產中心統包工程採購契約」（下稱系爭
    統包契約），由伊獨自負責系爭統包工程之規劃、設計，合
    先敘明。
  ㈠統包團隊共同標得系爭統包工程，並由中華工程公司擔任統
    包團隊之代表廠商，惟伊等在系爭統包工程中負責主辦之項
    目各有不同，依系爭統包工程共同投標協議書（下稱系爭共
    同投標協議書）之第2 條、第3 條之約定，伊主辦之項目為
    設計規劃，占系爭統包契約金額比率為2.5 ％；中華工程公
    司主辦建築工程，占系爭統包契約金額比率為83.87 ％；浩
    漢忠孝公司主辦自來水與機電工程，占系爭統包契約金額比
    率為13.63 ％。統包團隊請領工程款之方式，係由各成員各
    自出具發票及相關文件分別計價後交由代表廠商即中華工程
    公司統一向被告請領，由被告按各期進度分別辦理估驗並將
    各期估驗款項分別匯入統包團隊成員各自提供之帳戶，而非
    將工程款匯入統包團隊指定之聯名帳戶，亦即系爭共同投標
    協議書第6 條第1 項規範僅係用以簡化請款作業流程，而約
    定由代表廠商負責彙整統包團隊各成員相關請款文件及單據
    ，併代為請領預付款、估驗計價款及保留款等工程款項，絕
    無限制僅有代表廠商有請求被告給付工程款之權利。又依統
    包團隊歷來請領各期估驗計價款時所提出之工程計價請領明
    細表、計價請領簽認單之記載，可知統包團隊各成員依系爭
    統包契約施作之工程項目均可獨立劃分及累計估驗計價金額
    ，故統包團隊於施作後得請求被告給付之工程款債權，性質
    上屬可分之金錢債權，依民法第271 條規定，統包團隊各成
    員自得就各自施作部分，單獨起訴請求被告依約給付。況本
    件訴訟僅涉及伊負責規劃、設計作業之尾款請領爭議，與其
    餘統包團隊成員負責施作之項目及成果無涉，足認本件訴訟
    標的或判決結果對於統包團隊各成員尚無合一確定之必要，
    是以自不生未同列其他團隊成員為共同原告即有當事人適格
    要件欠缺之問題。
  ㈡伊已依約完成系爭統包工程之規劃、設計作業，而系爭統包
    工程復經被告於101 年2 月15日完成驗收，則依系爭統包契
    約第5 條第1 項第㈡款第⒈目：「第四期：全案工程完成驗
    收（含接水接電、設備完成試車合格）後，一次付清規劃設
    計費之尾款。」之約定，伊原得請求被告給付如附表所示項
    目之規劃設計費總計2,589 萬5,501 元，因被告前已給付伊
    1,986 萬6,365 元，另因系爭統包工程尚衍生變更設計後建
    物樓層高度落差之爭議，該部分所涉工程款之請領已由統包
    團隊另行起訴，故就樓高差減價部分爭議所涉之32萬8,268
    元設計費暨加計5%之加值營業稅1 萬6,413 元，亦應自本件
    伊請求之尾款金額中扣除，總計被告尚應給付伊568 萬4,45
    5 元（2,589 萬5,501 元－1,986 萬6,365 元－32萬8,268
    元－1 萬6,413 元=568萬4,455 元）。詎被告竟以伊在細部
    設計階段逾期完工107 日為由，扣留逾期違約金213 萬8,70
    4 元，惟系爭統包工程於伊完成規劃及初步設計作業後，將
    相關設計、規劃圖說文件送請臺北市政府都市發展局召開都
    市設計審議委員會（下稱都審委員會）審議時，經該會於第
    147 次會議中決議系爭統包工程得不受20公尺樓層高度限制
    ，全案設計朝建築物量體高度拉高、棟數減少、建物主體向
    內移設等方向重新配置設計（下稱系爭都審決議），嗣伊因
    應系爭都審決議內容提出都市設計審議修正報告書後，都審
    委員會始於95年6 月5 日准予核備。因系爭都審決議同意將
    系爭統包工程之建物樓層高度放寬至30公尺，而依91年7 月
    之「國軍年度施政工作計畫與預算編製作業規定」，樓層增
    高將導致建物單位造價成本升高，進而影響系爭統包工程之
    契約總價，是以在系爭統包契約依原計畫預算編訂之契約總
    價下，顯不可能達成設計、規劃、施作系爭統包工程使用單
    位即訴外人國防醫學院三軍總醫院（下稱三軍總醫院）所需
    求400 戶職務官舍之目標。為此，伊雖於95年6 月28日起開
    始提出部分第1 版細部設計內容送請被告委任之專案管理單
    位即訴外人台灣世曦工程顧問股份有限公司（下稱世曦公司
    ，原為中華顧問工程司）審查，惟被告及世曦公司於95年8
    月份工作月報會議中即要求統包團隊就系爭統包工程預估經
    費超過原預算情事提出後續建議方案，致當時已完成之細部
    設計內容需重新檢討規模及量體，且在工程總價未修正及施
    作樓地板總面積未確定前，伊亦無法編製預算書、單價分析
    表及工程數量計算書等契約約定之細部設計文件。嗣因被告
    與世曦公司要求統包團隊提出建議方案以供日後辦理契約變
    更作業時之參考，兩造及世曦公司遂自95年10月23日起進行
    17次價金檢討協調會，歷次會議中兩造及世曦公司就減項、
    減量、減樓層、減棟數等多種方案會商評估，遲至96年4 月
    25日第13次價金檢討協調會召開後，被告始決定採取減少A
    區A1、A2兩棟共80戶，總戶數由400 戶變更為320 戶之方案
    ，並由世曦公司於同年月30日正式發函通知伊按該次會議結
    論修改全案細部設計圖說及編列預算。嗣被告雖同意進行都
    市設計審議作業期間不計工期210 日，伊提送第1 版至第7
    版之細部設計作業審查期間不計工期72日，及自96年5 月10
    日經使用單位三軍總醫院同意全案刪減80戶之方案後，至兩
    造於96年10月15日簽訂「國防部軍備局工程營產中心統包工
    程附約」（下稱系爭附約）之期間，伊修改細部設計圖說及
    編列預算之相關期程不計工期159 日，惟自兩造於95年10月
    23日召開第1 次價金檢討調會至96年4 月25日召開第13次價
    金檢討協調會，期間長達184 日，被告僅核定不計工期72日
    ，顯然忽略歷次召開價金檢討協調會之目的係在解決被告預
    算不足問題，蓋因應系爭都審決議結果，原訂施作400 戶之
    工程預算已超過契約總價，而被告不可能超額核定，致伊在
    兩造及世曦公司決議確定全案設計方向前，無法於預定時程
    完成細部設計工作，自屬不可歸責伊之事由所致。故依系爭
    統包契約第7 條第2 項第9 款、第10款約定，細部設計階段
    除前述被告已同意不計工期之日數外，自95年10月23日第1
    次價金檢討協調會召開之日起至伊提送第3 版細部設計圖說
    文件送審而於同年11月1 日遭退回修正，並於同年月10日重
    新提送第4 版細部設計圖說文件之期間共計18日、伊提送審
    查之第5 版細部設計圖說文件於95年12月8 日遭退回修正，
    至伊在96年2 月14日重新提送第6 版細部設計圖說文件之期
    間共68日，及伊提送審查之第6 版細部設計圖說文件於96年
    3 月9 日遭退回修正，至伊另於同年4 月25日重新提送第7
    版細部設計圖說文件之期間共47日，均應不計工期，總計伊
    認定不應計入工期之日數尚有133 日（18＋68＋47＝133 ）
    ，是以被告於98年7 月27日以備工建管字第0000000000號函
    謂細部設計階段因逾期107 日，並依系爭統包工程契約第17
    條第1 項規定，以規劃設計費總額千分之一乘以逾期日數，
    對伊扣罰逾期違約金213 萬8,704 元，實無所據。
  ㈢被告復以浩漢忠孝公司之工程款債權涉有債權讓與等爭訟、
    統包團隊成員各自主張得請領之工程款金額有所矛盾、系爭
    統包工程之總保固金5,114 萬3,703 元中應包含自伊得請領
    之規劃設計費尾款按比例提列之122 萬6,741 元，故伊就該
    部分保固金應循保固金返還程序請求為由，刻意扣留如附表
    所列金額之款項，惟伊於系爭統包工程中僅負責規劃、設計
    作業，施作內容明顯可與統包團隊其他成員負責工作項目有
    所區分，且浩漢忠孝公司就其工程款債權因債權讓與所生紛
    爭，亦與伊無涉，又中華工程公司業已出具系爭統包工程尾
    款確認書及切結書表明該工程之保固金全額均由該公司繳納
    ，於工程保固期滿後亦應由該公司領取退還之保固金，故保
    固金繳納及退還均與伊無涉之旨，而此部分事實亦經另案臺
    灣高等法院103 年度建上字第149 號判決（下稱另案高院確
    定判決）所認定，綜此足認統包團隊內部業已確認系爭統包
    工程結算尾款7,068 萬4,832 元扣除保固金5,114 萬3,703
    元後，其中屬於伊之規劃設計費尾款金額為354 萬5,751 元
    ，且上開規劃設計費尾款金額加計遭被告扣留之逾期違約金
    213 萬8,704 元，即為本件伊請求被告給付之金額，再對照
    附表所列被告已支付及未支付之工程項目，均與被告所提之
    被告北部地區工程營產處第32期工程估驗計價單總表、工程
    估驗計價單明細表總表及工程結算明細表總表等資料相符，
    故顯無被告指稱統包團隊成員各自主張之尾款金額有矛盾情
    事，爰依系爭統包契約第5 條第1 項第㈡款第⒈目之約定，
    請求被告如數給付規劃設計尾款等語，並聲明如主文第1 項
    所示。
二、被告則以：系爭共同投標協議書第4 條及第6 條分別約定：
    「各成員於得標後連帶負履行契約責任。」、「共同投標廠
    商同意契約價金依下列方式請領：由代表廠商檢具各成員分
    別出具之發票及相關文件向機關統一請領。」，且依政府採
    購法第25條規定，共同投標成員間係居於共同承攬人之地位
    ，共同承攬施作工程，並連帶對業主負履行契約之責任，共
    同投標成員內部彼此存在類似合夥之關係，而系爭統包工程
    既係由統包團隊共同投標、與被告共同簽訂系爭統包契約，
    並且應連帶對被告負履約責任，則基於系爭統包契約所生之
    權利義務關係，自應由統包團隊共同享有或負擔，是應認本
    件工程款給付之爭議，其訴訟標的對於統包團隊全體有合一
    確定之必要。準此，本件僅由原告單獨起訴請求伊給付工程
    款，而非由原告與中華工程公司、浩漢忠孝公司共同起訴，
    已有當事人不適格之違失。又系爭共同投標協議書所載統包
    團隊「各成員所占契約金額比率」，僅為兩造訂約時之估計
    ，而統包團隊各成員於各期實際施作之情形多不相同，是不
    論統包團隊成員間如何依其等各期施作進度分配得請領之工
    程款比例，均應由代表廠商中華工程公司統一向伊請領，此
    亦為契約明定避免衍生款項請領紛爭之方式，況系爭統包工
    程歷來均由中華工程公司檢具統包團隊各成員分別出具之發
    票及相關文件統一請領各期估驗計價款，而原告既非代表廠
    商，更未檢具統包團隊各成員分別出具之發票及相關文件，
    顯不符兩造以契約明定之請款方式，伊自得拒絕給付。系爭
    統包工程有關工程尾款請領及保固金返還之爭議，雖經另案
    高院確定判決認定系爭統包工程由中華工程公司出具2 紙發
    票所提列之保固金總額5,114 萬3,703 元與原告無涉，及該
    工程結算尾款扣除上開保留款後所剩之1,222 萬977 元，其
    中354 萬5,751 元為原告之規劃設計費尾款等情，惟程序上
    仍應依系爭共同投標協議書第6 條約定之方式，由中華工程
    公司檢具發票及相關文件代表請領，如此方可避免後續再衍
    生統包團隊各成員主張之工程尾款或請求返還之保固金金額
    發生矛盾之紛爭。至原告主張系爭統包工程於細部設計階段
    遭認定逾期107 日係不可歸責於其之事由所致，不應扣罰逾
    期違約金213 萬8,704 元部分，參酌世曦公司所提細部設計
    階段工期檢討報告及其所發103 年11月24日世曦建字第0000
    000000號函暨就工程設計階段之工期部分陳述意見書內容，
    其結論略謂：「（世曦公司所提）審查意見中僅電梯車廂淨
    尺寸與價金會議結論有關，但均未影響圖說變更，其餘審查
    意見均為統包團隊圖說整合成果不完整，非可歸責於他造當
    事人（即被告）。」，是系爭統包工程規劃設計階段逾期10
    7 日，實係因統包團隊未依「統包」意旨，將規劃設計階段
    之圖說詳加整合所致，且依世曦公司就系爭統包工程指派之
    專案經理即訴外人吳佰玲之證言可知，系爭統包工程原係以
    400 戶職務官舍之需求為設計，嗣後以原告所完成400 戶之
    細部設計成果直接在圖面上加註刪減其中兩棟80戶，即為細
    部設計之定案版本，故伊未先行辦理契約變更作業，對於原
    告進行細部設計作業影響不大。況伊已就「A 區80戶A1-A2
    棟廢棄設計費」及「因都審決議變更產生之工程設計費」各
    給付原告189 萬8,500 元及66萬339 元，則原告因統包團隊
    自身圖面整合不佳致生延宕，終遭計罰逾期違約金，猶執系
    爭都審決議變更原規劃設計方向，致其無從在契約總價未確
    定下完成細部設計作業，故工程遲延係不可歸責於其之詞請
    求返還逾期違約金，自無理由等語，資為抗辯。並聲明：（
    一）原告之訴駁回。（二）願供擔保，請准宣告免為假執行
    。
三、本件經兩造整理不爭執事項及爭點事項如下：（見本院104
    年10月2 日準備程序筆錄，本院並依論述需要，調整其順序
    或文字用語）
  ㈠兩造不爭執之事項：
  ⒈原告與中華工程公司、浩漢忠孝公司成立統包團隊共同投標
    被告發包之系爭統包工程，由中華工程公司為統包團隊之代
    表廠商。統包團隊於93年7 月6 日標得系爭統包工程，並於
    同年8 月17日與被告訂立系爭統包契約。（見本院卷一第13
    頁至第16頁反面、第88頁至第110 頁）
  ⒉統包團隊於93年6 月29日簽訂之系爭共同投標協議書，經公
    證為系爭統包契約之一部。（見本院卷一第55頁至第56頁）
  ⒊系爭統包工程之請款方式係統包團隊依據系爭共同投標協議
    書第6 條約定，由代表廠商即中華工程公司檢具統包團隊各
    成員分別出具之發票、工程計價請領明細表、統包工程計價
    請領簽認單等相關文件，向被告統一請領。被告於歷次辦理
    估驗計價作業確認當期應付款項後，依統包團隊成員各自請
    領之金額，分別匯入統包團隊成員各自提供之帳戶。（見本
    院卷一第24頁、第55頁、第57頁至第58頁反面、第77頁反面
    、卷二第131頁反面）
  ⒋系爭統包工程於100 年12月30日竣工，並於101 年2 月15日
    經被告驗收合格。（見本院卷一第17頁）
  ⒌依系爭統包契約「第7 條履約期限、二、規劃設計階段分項
    履約期限、㈣」之約定，原告應於93年12月10日前完成規劃
    與初步設計各分項工作（含修正），並經被告審定；依同條
    項㈤之約定，設計圖說「達百分之三十」相關圖算資料經被
    告審定後，原告收到被告書面通知日起，應於94年9 月20日
    前完成細部設計工作計畫（含修正），並經被告審定；另依
    同條項㈨約定：「因不可歸咎於乙方（即原告）之事由，致
    不能如期交件時，其所受延遲時間不計工期。」；同條項㈩
    約定：「以上所列時程不含甲方（即被告）審核、作業、舉
    辦說明會及乙方向甲方簡報之時間。乙方工作成果修正以2
    次為限，超出次數如非屬甲方因素之修正日數應計算工期」
    。又同條「第五項、工程延期、㈠契約履約期間」約定：「
    有下列非可歸責於乙方之情形，而需展延工期者，乙方應於
    事故發生或消失後7 日內，以書面向甲方申請展延工期，不
    計算逾期違約金。⒉因甲方要求辦理變更設計或增加工程數
    量。⒌其它非可歸責於乙方之情形，經甲方認定者。」（見
    本院卷一第94頁反面至第95頁）
  ⒍原告進行初步設計簽證作業共使用工期111 日，嗣經被告於
    95年5 月15日核定不計工期169 日。被告另於97年1 月31日
    以備工建管字第0000000000號函核定都市設計審議作業期間
    不計工期210 日、細部設計作業審查時間不計工期72日、系
    爭統包工程於96年5 月10日經使用單位三軍總醫院同意核備
    改採320 戶方案後之變更設計期間不計工期159 日，總計於
    都市設計審議及原告進行細部設計作業階段不計入工期441
    日。（見本院卷一第76頁、第77頁、第111 頁、第112 頁、
    第166 頁、第167 頁）
  ⒎原告自95年6 月間開始提送部分細部設計圖說，並於95年9
    月5 日起至96年4 月25日止共提送7 版之全案400 戶細部設
    計圖說。嗣三軍總醫院於96年5 月10日同意將原需求戶數40
    0 戶刪減80戶之方案，原告即續辦320 戶之細部設計作業（
    提送第8 、9 版細部設計圖說），直至96年9 月7 日經被告
    核定全案320 戶細部設計工作計畫。（見本院卷一第77頁、
    第167 頁、第188 頁）
  ⒏系爭統包工程原規劃之基本戶數需求為16棟共400 戶，各可
    建地樓層為每棟地下1 樓、地上5 至7 樓不等，建築物每層
    樓高為1 樓樓層樑下淨高度3.2 公尺、其餘樓層版下淨高度
    3 公尺。嗣經都審委員會於94年7 月21日召開第147 次會議
    ，並作成系爭都審決議，即系爭統包工程興建之建築物得不
    受20公尺限高之限制，建築物高度可達30公尺，全案朝量體
    高度拉高（地下1 樓，地上5 至10樓不等），棟數減少（16
    棟調整為12棟）、量體向內側移設之方向重新配置設計。後
    經原告依上開決議內容修正報告書，都審委員會另於95年3
    月23日召開第164 次會議，確認修正報告書符合系爭都審決
    議內容，並於95年6 月5 日准予核備。（見本院卷一第168
    頁至第170 頁、第174 頁）
  ⒐系爭統包工程因系爭都審決議放寬樓高限制等結果，致原建
    築物量體、總建築開挖面積及總土方量有所變動，被告遂另
    行招商辦理環境影響差異分析評估作業，並於95年6 月16日
    向環保署掛件申請，至96年5 月10日經行政院環保署同意核
    備。（見本院卷一第167 頁、第175頁、第191 頁）
  ⒑統包團隊與被告進行歷次價金檢討協調會之會議結論詳如行
    政院公共工程委員會履約爭議調解案件卷宗申證6 所載。（
    見外放之行政院公共工程委員會履約爭議調解案件可閱卷宗
    第26頁至第27頁反面，下稱調解卷）
  ⒒統包團隊與被告於96年10月15日就系爭統包工程之延展工期
    及因應系爭都審決議變更致衍生之契約差異事項，簽訂系爭
    附約，約定基本戶數需求自原決標之16棟（400 戶）減少為
    10棟（320 戶），可建地樓層調整為每棟地下1 樓、地上5
    至10樓不等，建築物每層樓高為1 樓樑下淨高度3 公尺、其
    餘樓層版下淨高度3 公尺，契約價金由9 億1,140 萬元變更
    為8 億7,464 萬2,999 元，其中320 戶之工程設計費議定為
    1,998 萬7,887 元，廢棄設計費議定為145 萬7,132 元。（
    見本院卷一第189 頁至第191 頁）
  ⒓系爭統包工程之保固金為5,114 萬3,703 元，中華工程公司
    曾於102 年12月19日函文記載：「本案需繳交5%保固金計5,
    114 萬3,703 元，本統包團隊同意由本案三家聯合承攬廠商
    剩餘工程尾款依承攬比例分別提列。」。中華工程公司另於
    102 年12月25日開立2 紙金額分別為1,158 萬9,811 元及3,
    955 萬3,892 元之發票予被告。（見本院卷一第78頁、卷二
    第24頁、第28頁、第29頁）
  ⒔系爭統包工程結算後之工程尾款為7,068 萬4,832 元，因該
    工程之保固金5,114 萬3,703 元係以尾款轉繳保固金方式繳
    納，而扣除保固金及全案（含設計、施工、驗收階段）履約
    逾期違約金732 萬152 元後，尾款總額為1,222 萬977 元。
    （見本院卷一第64頁至第65頁反面、第66頁反面至第67頁）
  ⒕被告尚未給付原告如附表所列之金額。（見本院卷二第127
    頁反面）
  ㈡本件爭點：
  ⒈原告單獨向被告起訴請求給付規劃設計費，當事人適格是否
    欠缺？
    ⑴統包團隊共同投標、承攬被告之系爭統包工程，並簽立系
      爭統包契約暨附約，依照系爭共同投標協議書第6 條約定
      ，承攬之工程款債權是可分或不可分之債？
    ⑵原告可否單獨起訴請求被告給付規劃設計費？
  ⒉原告請求被告給付如附表所示規劃設計費尾款568 萬4,455
    元，有無理由？
    ⑴原告請求給付規劃設計尾款，是否符合系爭統包契約第5
      條第1 項第㈡款第⒉目第⑶次目之約定？
    ⑵系爭統包工程於細部設計階段，得不計入工期之日數為何
      ？被告以原告於規劃設計階段逾期107 日曆天為由，依系
      爭統包契約第17條規定計罰逾期違約金213 萬8,704 元，
      有無理由？
四、本院之判斷：
  ㈠原告得單獨向被告起訴請求給付規劃設計費，當事人適格無
    欠缺。
  ⒈按營造業法第3 條第7 款規定：「聯合承攬：係指二家以上
    之綜合營造業共同承攬同一工程之契約行為」；政府採購法
    第25條第1 項、第2 項、第5 項亦明定：「機關得視個別採
    購之特性，於招標文件中規定允許一定家數內之廠商共同投
    標。」、「前項所稱共同投標，指二家以上之廠商共同具名
    投標，並於得標後共同具名簽約，連帶負履行採購契約之責
    ，以承攬工程或提供財物、勞務之行為。」、「共同投標廠
    商應於投標時檢附共同投標協議書」。查原告與中華工程公
    司、浩漢忠孝公司所成立之統包團隊共同投標取得被告發包
    之系爭統包工程，並與之簽訂系爭統包契約暨附約，由原告
    負責系爭統包工程之設計、規劃作業；中華工程公司主辦建
    築工程；浩漢忠孝公司則負責自來水與機電工程之施作，有
    系爭共同投標協議書可參（見本院卷一第55頁），堪認原告
    與中華工程公司、浩漢忠孝公司所為係異業結合共同投標之
    聯合承攬行為。至聯合承攬人即統包團隊對業主即被告之工
    程款債權係可分或不可分，應視統包團隊與被告間是否有所
    約定，或有無法律另有規定情形，此觀諸民法第271 條所定
    「數人負同一債務或有同一債權，而其給付可分者，除法律
    另有規定或契約另有訂定外，應各平均分擔或分受之；其給
    付本不可分而變為可分者亦同。」即明。
  ⒉次查兩造對於系爭統包契約第1 條約定契約文件包含系爭共
    同投標協議書乙節，並不爭執（詳前開三、㈠兩造不爭執之
    事項、⒉所示）。觀諸系爭共同投標協議書第1 條、第3 條
    、第4 條、第6 條明揭：「共同投標廠商同意由中華工程公
    司為代表廠商，並以代表廠商之負責人為代表人，負責與機
    關意見之聯繫」、「各成員所占契約金額比率：第一成員（
    即中華工程公司）：83.87%、第二成員（即原告）：2.5%、
    第三成員（即浩漢忠孝公司）：13.63 % 。」、「各成員於
    得標後連帶負履行契約責任。」、「共同投標廠商同意契約
    價金依下列方式請領：由代表廠商檢具各成員分別出具之發
    票及相關文件向機關統一請領。」等內容（見本院卷第55頁
    ），對照被告所提之系爭統包工程第32期工程計價請領明細
    表、工程計價請領簽認單、中華工程公司出具之第32期工程
    款發票（見本院卷一第57頁至第58頁反面），及系爭統包工
    程專案管理單位世曦公司檢送被告核復之系爭統包工程第32
    期工程估驗計價單明細表總表（見本院卷一第60頁至第63頁
    ）等資料，足以推知系爭統包工程之第32期估驗計價款，係
    由統包團隊成員共同確認原告負責之「工程設計當期計價金
    額」、中華工程公司施作之「土木建築工程當期計價金額」
    及浩漢忠孝公司承攬之「水電工程當期計價金額」分別為0
    元、1,615 萬2,000 元及0 元，並由中華工程公司檢送發票
    及工程計價請領簽認單等文件予被告，而被告則依該請求計
    價之金額如數支付中華工程公司之事實。再參酌被告於98年
    7 月27日所發備工建管字第0000000000號函及附有記載被告
    支付系爭統包工程第2 期規劃設計費635 萬7,240 元予原告
    ，原告於98年6 月10日出具同額收據予被告之支出憑證黏存
    單（第1 號）等文件（見本院卷一第24頁、第77頁反面），
    及兩造均不爭執被告實際給付系爭統包工程各期款項之方式
    ，係將各期依統包團隊各成員分別出具發票或相關文件委由
    代表廠商中華工程公司向被告請求估驗計價之款項，匯入各
    成員指定之帳戶，並非匯入統包團隊之聯名帳戶（詳前開三
    、㈠兩造不爭執之事項、⒊所示），再由統包團隊成員於收
    款後個別出具收據予被告等情，綜此堪認統包團隊各成員與
    被告約定，除就契約履行負連帶責任外，各成員係就其等主
    辦項目分別出具發票及相關文件分別列計各期請求估驗計價
    金額，僅係由代表廠商中華工程公司統一向被告請領，由此
    亦足推論統包團隊各成員對被告之工程款債權乃係明確可分
    。
  ⒊被告雖辯稱其無法區分系爭統包工程是由何人施作、設計，
    因而無法釐清工程款應給付予何人，且統包團隊各成員間主
    張之工程尾款金額有所矛盾，如統包團隊各成員未一同起訴
    ，不僅與系爭共同投標協議書第6 條約定之請款方式有違，
    其亦無從確認統包團隊各成員對於系爭統包工程結算後扣除
    全案保固金之尾款1,222 萬977 元所占金額比例為何云云。
    然查，依系爭統包工程專案經理即證人吳佰玲證述：「如果
    從計價請領明細表上面有分列土木建築工程、工程設計、水
    電工程每一期計價累計金額，最後會有一個總計的金額…如
    果原告認為有尚未取得的設計費用，只要他們統包團隊成員
    能夠確認，並據此向被告請求，應該是可以的」、「基本上
    我們尊重統包團隊提出的內容，對我們而言只要那一期工期
    合計的計價金額依照設計、施工之工程進度的比例是正確的
    ，我們就會同意進行估驗計價。設計、施工的工程進度是分
    別依照統包團隊成員所提出的資料來認定，例如施工的部分
    是針對土木及機電工程各自的進度比例作認定。」等語（見
    本院104 年8 月7 日準備程序筆錄，本院卷二第120 頁、第
    122 頁反面），足知統包團隊各成員於系爭統包工程負責施
    作之工程項目明確可分，且世曦公司亦係就統包團隊各成員
    分別提出之各期施作進度辦理設計及施工部分之歷次估驗計
    價作業，故系爭統包工程既僅由原告負責規劃、設計作業，
    則原告得向被告請求給付規劃、設計工作之對價，要無疑義
    。又依中華工程公司於104 年4 月30日以（104 ）中工法字
    第000000000000號函說明有關系爭統包工程結算尾款餘額1,
    222 萬977 元中，屬於原告之工程尾款金額為354 萬5,751
    元，隨函並檢附原告與中華工程公司簽立之尾款確認書等節
    （見本院卷二第24頁、第25頁、第30頁），堪認統包團隊成
    員彼此間就系爭工程結算尾款各自得請求之尾款金額已可相
    互確認，再據以向被告請求給付。而依前開尾款確認書所確
    認屬於原告得請領之規劃設計費尾款金額354 萬5,751 元，
    如加計被告主張原告遲延履約所扣留之逾期違約金213 萬8,
    704 元，適為原告本件請求之規劃設計費尾款總額568 萬4,
    455 元【計算式：3,545,751 ＋2,138,704 ＝5,684,455 】
    ，又被告對於原告就附表所列各請求項目對照被告所提被告
    北部地區工程營產處第32期估驗計價單明細表總表、工程結
    算明細表總表後列表說明之內容（見本院卷二第94頁正、反
    面），已表示原告請求各項目之金額與上開估驗計價或結算
    資料內容相符（見本院卷二第114 頁），甚且中華工程公司
    與原告依前開尾款確認書確認屬於原告規劃設計費之尾款金
    額，亦經另案高院確定判決於判決理由中加以確認，此有另
    案高院確定判決在卷可按（見本院卷二第272 頁、第273 頁
    正、反面），而被告對於另案高院確定判決之認定結果復未
    爭執（見本院卷三第23頁反面），益證被告前揭所稱無從確
    認係何人負責系爭統包工程之設計作業、無從確認原告得請
    求之尾款金額或統包團隊各成員間彼此主張之尾款金額有所
    矛盾云云，俱無值採。
  ⒋共同投標廠商全體，均為採購契約主體，而為共同承攬人之
    一，縱非共同投標採購契約之代表廠商，惟各成員之主辦工
    程項目、所占契約金額比率，併價金請（受）領之方式、項
    目及金額等，既已載明於共同投標協議書，堪認各成員基於
    共同投標採購契約所生之工程款債權，明確而適於執行，各
    成員非不得本於契約法律關係，逕向採購契約之對造機關，
    請求給付主辦項目之工程款。是以本件兩造間約定統包團隊
    各成員對被告之工程款債權屬可分之債，則原告自得依系爭
    統包契約之法律關係，單獨對被告起訴請求給付工程款，因
    此，被告抗辯原告單獨起訴欠缺當事人適格云云，亦非可採
    。
  ㈡原告得請求被告給付如附表所示規劃設計費尾款568 萬4,45
    5 元。
  ⒈系爭統包工程業於101 年2 月15日經被告驗收完畢，有工程
    結算驗收證明書存卷可參（見本院卷一第17頁），而被告迄
    未給付如附表所示款項予原告，為兩造所不爭執（詳前述三
    、㈠兩造不爭執之事項、⒕所示），又原告請求如附表所示
    尾款金額扣除遭被告扣留之逾期違約金後，所剩354 萬5,75
    1 元為經統包團隊成員確認後原告就系爭統包工程結算尾款
    1,222 萬977 元中得請求之規劃設計費尾款金額，亦如前述
    ，則原告依系爭統包工程契約第5 條第1 項第㈡款第⒈目之
    約定，請求被告給付規劃設計費尾款354 萬5,751 元部分，
    當屬有據。至被告雖辯稱原告未委由代表廠商檢具發票及相
    關文件統一向其請領尾款，與系爭統包契約及系爭共同投標
    協議書之約定內容不符云云，惟本件統包團隊各成員對被告
    之工程款債權係可分之債，得由原告單獨起訴請求被告給付
    ，業已詳論如前，且原告就其請求之規劃設計費尾款是否出
    具統一發票，核與被告依約所負給付工程款之義務並不具對
    待給付關係，故被告執此為由拒絕給付前揭規劃設計費尾款
    ，洵無足取。
  ⒉被告固另提出系爭統包工程之保固年限總表、保固年限及金
    額統計表及中華工程公司於102 年12月19日所出具（102 ）
    中工建字第000000000000號函，聲稱系爭統包工程全案之保
    固金5,114 萬3,703 元係由統包團隊之成員依承攬比例分別
    提列，故上開保固金應包含依原告於系爭統包工程得請領之
    全部規劃設計費2,453 萬4,822 元提列5%之保固金122 萬6,
    741 元（見本院卷一第78頁、卷二第128 頁反面、第165 頁
    反面至第167 頁），原告就此部分應依約循保固金返還程序
    請求退還云云。然查，被告前揭主張為原告所否認，並謂系
    爭統包工程係於101 年2 月15日經被告驗收合格，而原告依
    保固年限規定表之規定，僅須負1 年之保固責任，故自系爭
    統包工程驗收完畢之日起算1 年，原告之保固期間已在102
    年2 月14日屆滿，是以在中華工程公司於同年12月25日以全
    案工程結算尾款轉繳保固金時，因原告已免負保固責任，故
    未就其得請領之工程款項按承攬比例提列保固金，另提出被
    告之保固年限規定表為憑（見本院卷三第12頁），堪信原告
    所述，尚非無稽。而依被告所提原告在103 年2 月26日出具
    之切結書，其上已明載系爭統包工程之保固金係由中華工程
    公司繳納，故於保固期滿後，應由中華工程公司向被告請求
    退還保固金，原告對於保固金不得主張任何權利之旨（見本
    院卷一第78頁反面），此與中華工程公司檢附原告及浩漢忠
    孝公司簽立之切結書2 份、中華工程公司於102 年12月25日
    開立之工程尾款發票2 紙（金額分別為1,158 萬9,811 元、
    3,955 萬3,892 元）及原告與中華工程公司共同簽立之尾款
    確認書，於104 年4 月30日函覆本院系爭統包工程之保固金
    5,114 萬3,703 元係由該公司與浩漢忠孝公司共同提列（金
    額分別為3,955 萬3,892 元、1,158 萬9,811 元），浩漢忠
    孝公司並同意由中華工程公司於保固期間屆滿後向被告請求
    退還1,158 萬9,811 元之保固金，而原告並未自其得請領之
    工程款提列任何金額充作保固金等情，及中華工程公司於10
    5 年3 月1 日復行出具切結書之文義均相符（見本院卷二第
    24頁至第30頁、卷三第21頁），又參酌證人吳佰玲證稱系爭
    統包工程之保固款是由中華工程公司提列之內容（見本院卷
    二第120 頁）、中華工程公司以保固期間屆至且保固缺失均
    已改善完成為由，向本院另訴請求被告退還保固金3,955 萬
    3,892 元（見本院卷二第277 頁至第280 頁）及另案高院確
    定判決於判決理由中亦認定系爭統包工程之全案保固金並未
    包含自原告按承攬比例提列之部分工程款項等節（見本院卷
    二第273 頁反面），足認系爭統包工程之保固金係由中華工
    程公司單獨繳納，原告之規劃設計費尾款並未提列其中之事
    實，縱然全案保固金之數額係以系爭統包契約總金額之5%列
    計（見本院卷一第93頁），惟保固金數額之計算與實際繳納
    保固金並得請求返還之人分屬二事，且系爭統包契約亦無統
    包團隊各成員均須提列部分工程款充作保固金之明文約定，
    是以本件統包團隊達成共識由中華工程公司負責繳納按系爭
    統包契約總價乘以5%計算之保固金，尚與契約之約定無違。
    準此，原告既已切結系爭統包工程之保固金與其無涉且聲明
    不再主張任何權利，則被告猶執系爭統包契約第5 條第1 項
    第㈡款第⒉目第⑶次目之約定，辯稱原告就其請求尾款中之
    122 萬6,741 元屬於保固金性質，應另循保固金退還程序辦
    理云云，殊難憑採。
  ⒊就原告請求如附表所示之金額568 萬4,455 元，扣除本院前
    揭認定原告得請求被告給付之規劃設計費尾款354 萬5,751
    元，所剩213 萬8,704 元部分，為被告抗辯係因原告於細部
    設計階段遲延工期107 日而依約扣罰之逾期違約金，其除援
    引原告所提世曦公司提送被告核備之細部設計階段工期檢討
    報告外，另提出世曦公司於103 年11月24日以世曦建字第00
    00000000號函及隨函檢附之「事實及理由之陳述意見- 有關
    工程設計階段之工期部分」、系爭統包契約暨附約、系爭統
    包工程細部設計階段履約工期檢討網圖、系爭統包工程都市
    設計審議履約期限檢討說明、系爭統包工程專案管理（含監
    造）送審文件處理狀況彙總表、工程圖說/ 技術文件審查意
    見表等資料（見本院卷一第128 頁至第130 頁反面、第156
    頁至第191 頁），欲以此證明前揭細部設計階段被告認定遲
    延之工期107 日，均係因原告未將規劃設計之圖說詳加整合
    所致，故屬可歸責於原告之事由，應由原告承擔逾期完成細
    部設計作業之違約責任等事實，然此情為原告所否認，並主
    張於系爭都審決議後，原定規劃設計內容需再重新檢討規模
    及量體，且依系爭都審決議內容執行原定設計方案，將導致
    建物造價成本增加，進而超出原定契約總價，而因被告不同
    意提高施作系爭統包工程之總預算，故兩造與世曦公司因而
    召開17次價金檢討協調會以確認如何在原定契約總價下提出
    既符合系爭都審決議內容，亦合乎使用單位三軍總醫院需求
    之細部設計方案，而在第13次價金檢討協調會召開前，系爭
    統包契約之總價及應施作之樓地板面積均處於未確定狀態，
    其無從進行有實益之細部設計作業，故應認自95年10月23日
    召開第1 次價金檢討協調會之日起，其歷次提送世曦公司審
    查細部設計圖說之期間及該公司審查後退回要求其修正之期
    間，均應不計入工期，總計應再有133 日不計工期，被告所
    指其遲延完成細部設計作業之期間共107 日，實非可歸責於
    其之事由所致等語。經查：
  ⑴原告自95年9 月5 日起至96年4 月25日止共提送7 版之全案
    400 戶細部設計圖說，嗣於96年4 月30日經世曦公司正式發
    函通知及三軍總醫院於同年5 月10日同意將原需求戶數400
    戶刪減80戶之方案後，原告乃續辦320 戶之細部設計作業（
    提送第8 、9 版細部設計圖說），直至96年9 月7 日經被告
    核定全案320 戶細部設計工作計畫；被告就規劃、設計階段
    核定不計工期之具體情形及原告因應系爭都審決議而提出修
    正報告書，將全案細部設計朝建物量體高度拉高（地下1 樓
    ，地上5 至10樓不等），棟數減少（16棟調整為12棟）、量
    體向內側移設之方向重新配置設計等情節，詳如前揭三、㈠
    兩造不爭執之事項、⒍、⒎、⒏所示。而細繹世曦公司所製
    作之「本案都市設計審議履約期間檢討說明」文件，可知因
    都審委員會考量附近居民抗爭訴求，故作成系爭都審決議放
    寬樓高20公尺之限制，使系爭統包工程施作之建物高度可達
    30公尺，又因建物量體拉高，因而可縮減整體建物棟數，進
    而影響原建物之配置設計（見本院卷一第170 頁），再酌之
    系爭附約第二條第㈠項內容提及依91年7 月「國軍年度施政
    工作計畫與預算編製作業規定」之標準，因應系爭都審決議
    而將建物量體拉高之造價成本相較於決標時原單位造價落差
    級數為12.5% （見本院卷一第189 頁），及原告所提系爭統
    包工程95年8 月份工作月報會議記錄、95年10月份工作協調
    會會議記錄，其中顯示被告要求統包團隊因應系爭都審決議
    ，就原設計400 戶職務官舍方案可能超出被告之工程預算乙
    事研討後續建議、兩造及世曦公司遂定於95年10月23日起就
    系爭附約之簽訂召開價金檢討協調會等情（見本院卷一第11
    8 頁正、反面、第121 頁反面至第123 頁），應足認定原告
    所稱因系爭都審決議結果實質上造成依原設計方案執行，將
    使系爭統包工程施作成本超出系爭統包契約之總價，被告復
    要求在原定契約總價下調整細部設計方案，故其在96年4 月
    30日接獲世曦公司通知確定改採總戶數320 戶之設計方案前
    ，無從確定全案細部設計之方向，且於系爭附約完成議價及
    簽約前，系爭統包契約之總價亦處於浮動狀態等節，尚非子
    虛。
  ⑵觀諸系爭統包契約第二條履約標的、五、㈠規劃設計、⒍細
    部設計之規定，可知細部設計階段原告應完成之工作至少包
    含繪製細部設計圖（包含工程平面、立面、剖面、結構等詳
    圖、水電、消防、空調、電信…等及建築法規所規定之必要
    建築物設備，及其他道路、排水、道路照明等之附屬工程相
    關詳圖，以及必要之相關管線配合拆遷設計）、設計計算書
    （提供結構、水電、空調、消防、排水、環工等之設計計算
    書）、施工規範及器材規範、工程數量計算書（各項計算式
    應詳細註明尺寸及使用位置）、預算書及單價分析表、水土
    保持計畫等（見本院卷一第90頁反面），對照世曦公司於95
    年10月份工作協調會所提出辦理系爭附約訂定價金檢討資料
    ，其中㈢、⒊記載都審後細設成果為：樓高由5 至7 樓調整
    為5 至10樓、棟數由16棟調整為12棟、增作基樁、重新設計
    廢棄不用部分之規劃設計費，且依該份資料所引系爭統包契
    約條款足知兩造為因應系爭都審決議，已確立將來以簽訂系
    爭附約之方式辦理契約變更作業（見本院卷一第124 頁反面
    至第126 頁）。而參酌原擔任原告工務部經理之證人梁有忠
    到庭所證：「都審委員有採納居民之意見，因此把樓層高度
    放寬至30公尺…因為被告有一個國軍的造價標準…樓層高度
    拉高興建成本就會變高，成本變高之後會將原來的預算提高
    ，我印象中樓層拉高一級造價增加12.5% ，所以原來的9 億
    多的預算應該要增加到11億元…是因為系爭都審決議決定拉
    高樓層連帶影響造價成本，而被告預算不足，且超出10億元
    還要另外走行政程序，所以最終議定就減為320 戶。」、「
    因為統包工程得標時，總價及樓地板面積是固定的，這兩項
    其中一個變更都叫契約變更，本件兩項都有變更，所以一定
    要辦理契約變更作業，當時我們有要求先做契約變更之後，
    重新議定總價及樓地板面積，再依據契約變更後議定之事項
    去進行後續的細部設計作業」、「我們是討論到第13次才定
    案…直到第13次會議才決定將最靠近擋土牆的2 棟建築刪掉
    。」、「依照統包契約之規定，細部設計要完成的事項包含
    提出工程的預算書、工程數量表、單價分析表，如果價金沒
    有確定就無法做工程預算書，且在圖說上面是否要減量、減
    戶、減棟，會涉及到其他機電工程系統之問題，例如機電容
    量等就會有所改變，被告在未確定要減量或是減戶之前，原
    告所為的設計都是無效的設計。」、「細部設計有建築圖、
    機電圖等，但是世曦公司的審查比較侷限於景觀工程，當時
    減項、減量也會影響景觀工程…即使依照世曦公司歷次審查
    的結論去修正圖說，也是無法提出預算完備所有的細部設計
    內容。我們認為應該將主要的設計方向確定下來，歷次審查
    的修正在主要方案確定下來之前均無實益」等語（見本院10
    4 年8 月7 日準備程序筆錄，卷二第117 頁至第119 頁），
    經核與前揭系爭附約條文第二條第㈠項內容、工作月報或協
    調會會議記錄結論並無二致，且與證人吳佰玲證稱：「當初
    送都審的時候的確是400 戶沒有錯…但是因為有居民抗爭，
    所以要求我們的配置要後退，最後定案的結論就是將樓層拉
    高，棟數減少，都審核備之後就做細部設計，第二次細部設
    計成果提出後，統包團隊經估算後，這樣量體的工程費約12
    億元，已經超出當初原定10億元的預算」、「價金檢討主要
    是討論增加的2 億元的成本部分，7 樓至10樓的單位造價比
    較高，比1 至7 樓單位造價超出12.5 %，如果依照高樓層的
    造價處理，用原定400 戶去估算就會超出10億元的預算，所
    以歷次價金檢討會議就是針對如何刪減進行討論，最後決定
    刪減2 棟80戶的方案。」、「基本上原告還是以400 戶去設
    計，再用400 戶的成果去刪減其中的80戶，在320 戶確定下
    來之後，確實會有設計單位主張部分機電容量等應該相應調
    整的問題，所以我們另外在附約就有議定要調整的部分。」
    等內容大致相符（見本院卷二第120 頁反面至第121 頁反面
    ），由此應足認定在系爭都審決議後，因契約總價及全案設
    計方向均有所改變，兩造亦已達成共識應辦理契約變更作業
    ，則在正式確定採行何種設計方案既可符合系爭都審決議意
    旨，亦不會超出系爭統包契約總價前，原告縱使依約提送以
    400 戶為設計主軸之細部設計圖說文件供世曦公司審查並依
    審查意見作修正，惟其依約提送及修正以400 戶職務官舍為
    基礎之設計結果顯與系爭統包工程最終決議採行縮減為320
    戶設計之方案有別，遑論刪減80戶後之方案對於工程預算之
    編列、數量之計算乃至於契約總價均有莫大影響，則原告所
    稱系爭統包工程如先完成契約變更程序，並於該程序中一併
    議定工期，其即可具體掌握全案設計方向並續行細部設計作
    業，兩造亦可清楚釐清逾期責任歸屬乙節，較為可採。
  ⑶被告固執前揭送審文件處理狀況彙總表、工程圖說/ 技術文
    件審查意見表指摘原告歷次提送審查之細部設計圖說多有未
    依審查意見修正之缺失，且證人吳佰玲於作證時迭稱：「為
    了減少統包團隊變更的負擔，就是以400 戶的細設成果在圖
    面上加註不包含A1及A2棟」、「我們為了節省時間，所以採
    用同步作業，一邊由原告細設，我們負責審查細設的成果，
    一邊進行價金檢討會議，討論如何刪減價金的方案，我們基
    本上不會動到建築主體的配置，所以原告還是可以進行細部
    設計作業，待原告完成細設圖說後，直接在圖面上加註刪減
    的2 棟樓房，所以並沒有不能同步進行細部設計作業的問題
    。」、「我們一直跟統包團隊溝通，我們一直建議照原來40
    0 戶去設計，如果有涉及到價金檢討事項再另行辦理議價的
    程序，最後就是直接在400 戶的圖說上刪減2 棟樓房，不會
    有原告主張主方案未確定、規格變更或材質變更影響到工期
    之問題」等語（見本院卷二第120 頁反面至第121 頁、第12
    2 頁）。惟細觀兩造均不爭執之原告於提起本件訴訟前，向
    行政院公共工程委員會採購申訴審議委員會聲請履約爭議調
    解時所提出其整理之價金檢討協調會一覽表（見調解卷第26
    頁至第28頁反面），可知統包團隊於歷次價金檢討協調會中
    已多次陳明被告應先依約辦理契約變更之意見，然未獲被告
    及世曦公司所採納，而歷次會議中兩造及世曦公司曾提出數
    種刪減項、量、戶數、棟數之方案以試算價金，再依試算結
    果修正方案內容或調整業主需求，直至96年4 月25日召開第
    13次價金檢討協調會時方確定採行刪減A1、A2兩棟80戶之方
    案，惟原告為細部設計時尚須預留管線以因應將來可能增建
    官舍之需求等情，再佐以證人吳佰玲前開所證，適足證明於
    系爭都審決議後，為解決依原設計方案施作工程將衍生工程
    成本超出原定契約總價之問題，兩造及世曦公司持續透過召
    開價金檢討協調會之方式以商討全案設計之方向及辦理契約
    變更即簽訂系爭附約之前置作業，則在第13次價金檢討協調
    會召開前即要求原告在契約總價及設計方向均不明確之情形
    下，編列工程預算或計算詳細數量，實屬苛求。況證人吳佰
    玲亦證：「如果照正常的作法是確定幾戶之後再進行細部設
    計，這是最單純的，但是本件是統包團隊需限期完工，都審
    後我們同步進行細部設計、環境差異分析及價金檢討，同步
    進行就是希望整個工期可以掌握住，正常的流程是都審變更
    之後，要進行環差分析再進行細部設計，但是因為需限期完
    工，所以專案管理單位與兩造達成共識要同步進行，故確實
    是會有證人梁有忠所述可能某些部分會有重複設計的問題。
    」等語（見本院卷二第123 頁反面），由此足知系爭統包工
    程並未採行一般工程中先辦理契約變更，再依變更結果續辦
    細部設計作業或施工之流程，惟證人吳佰玲或被告均未具體
    說明基於減少不同廠商施作介面之整合或圖面溝通問題、掌
    握整體完工期程、提升採購效率等目的而採行將設計及施工
    合併招標及簽約執行之統包工程模式，即表示因系爭都審決
    議導致須辦理契約變更作業，而在辦理契約變更前所耗費履
    約期間之不利益即應由統包團隊自行承擔之理由，且被告亦
    未證明原告未確實依世曦公司就歷次細部設計圖說之審查意
    見修正細部設計內容，即為導致兩造須召開數次價金檢討協
    調會，乃至簽訂系爭附約以解決原告設計不當或整合圖說不
    完備等問題之主因，是以依被告前開所辯或證人吳佰玲之證
    言，無異形成被告或世曦公司先是不同意統包團隊提出依約
    辦理契約變更作業之要求，迫使原告仍須依約進行細部設計
    作業，嗣於原告依審查意見為數次修正後，提送以400 戶為
    設計主軸之細部設計圖說，最終仍不符第13次價金檢討協調
    會會議結論採行以全案320 戶為定案設計方向，致原告須另
    依320 戶定案方向重新進行細部設計之結果，則此種作業方
    式是否確實合乎被告採行統包模式以縮短整體工期之原意，
    實有可疑。又倘依證人吳佰玲所述採行320 戶之設計方案僅
    須於原告提送之400 戶細部設計圖說上直接加註刪減2 棟共
    80戶即可完成，則何以無論觀之世曦公司提送被告核備之細
    部設計階段檢討報告或被告所提會驗結果報告單，其中分別
    認定自96年5 月1 日（世曦公司建議版本）及96年5 月10日
    （被告核定版本）起即不計入原告辦理細部設計作業之工期
    ，且依被告所提兩造及世曦公司於簽訂系爭附約前往來函文
    、為簽訂系爭附約之投標須知、議價/ 決標紀錄、變更設計
    申請書及工程變更設計明細表與單價分析表等資料（見本院
    卷二第210 頁至第232 頁、第234 頁至第247 頁反面），亦
    可推知原定契約內容變更之幅度非微，當非直接自全案400
    戶之細部設計圖說文件上加註刪減80戶即可完成，由此益證
    證人梁有忠所證：「議訂320 戶之方案之後，被告有給全部
    的工期，如此可知，前面所進行的作業時間都應扣除，因為
    320 戶方案的擬定是主要徑。」、「400 戶變成320 戶…這
    涉及到其他結構及機電系統的連結，連帶工程施作的基樁都
    會受影響，採用320 戶的方案確定後，原告還花了2 、3 個
    月的時間進行設計，機電系統是否要保留在確定320 戶的方
    案之後還討論很多次，這些都會導致價金不確定，價金是細
    部設計要核定的要項」等語（見本院卷二第123 頁正、反面
    ），確值採信。
  ⑷綜上，系爭統包工程之細部設計作業，因系爭都審決議內容
    而需相應調整，而在兩造於達成採行刪減2 棟80戶之定案設
    計方向前，原告難以就未臻明確之契約總價編列預算、估算
    工程數量及繪製符合系爭都審決議及被告需求之細部設計圖
    說文件，已如前述，則應認因系爭都審決議導致兩造召開第
    1 次至第13次價金檢討協調會商討全案設計方向及試算契約
    價金之期間，應屬不可歸責於原告所增加之履約期程，則依
    系爭統包契約第7 條第2 項第㈨款、第款、同條第5 項第
    ㈠款第2 目、第5 目所定「因不可歸咎於乙方（即統包團隊
    ，依該條款內容應指原告）之事由，致不能如期交件時，其
    所受延遲時間不計工期。」、「規劃設計需求辦理變更，其
    項目或數量有增減時，履約期限得由雙方視實際需要議訂增
    減之。」、「契約履約期間，有下列非可歸責於乙方之情形
    ，而需展延工期者，乙方應於事故發生或消滅後7 日內，以
    書面向甲方申請展延工期，不計算逾期違約金：⒉因甲方要
    求辦理變更設計或增加工程數量。⒌其他非可歸責於乙方之
    情形，經甲方認定者。」等規範（見本院卷一第94頁反面至
    第95頁），應認原告主張自95年10月23日召開第1 次價金協
    調會起至95年11月10日其修正後提送第4 版細部設計圖說、
    自95年12月8 日第5 版細部設計圖說遭退回修正至其在96年
    2 月14日提送第6 版細部設計圖說、自96年3 月9 日第6 版
    細部設計圖說遭退回修正至其在96年4 月25日提送第7 版細
    部設計圖說，共計133 日亦應不計入工期，尚屬有據，則被
    告未斟酌前揭細部設計階段不可歸責於原告致履約發生遲延
    情事，而核定原告逾期107 日完成細部設計工作，並依系爭
    統包契約第17條第1 項規定，自原告之規劃設計費中扣留21
    3 萬8,704 元為逾期違約金，自無理由，是以原告就該遭被
    告扣留之213 萬8,704 元規劃設計費，應得連同前述規劃設
    計費尾款354 萬5,751 元部分一併請求被告給付，即被告應
    給付原告共計568 萬4,455 元【2,138,704 元＋3,545,751
    元＝5,684,455 元】。
五、綜上所述，原告本於前揭原因事實，主張系爭統包工程業已
    完工且經驗收合格，原告復無逾期完成細部設計作業情事，
    而依系爭統包契約第5 條第1 項第㈡款第⒈目約定，請求被
    告給付如附表所示工作項目暨稅捐之規劃設計費尾款共計56
    8 萬4,455 元，及自起訴狀繕本送達被告翌日即103 年8 月
    12日起（被告係於103 年8 月11日收受起訴狀繕本，見本院
    卷一第42頁）至清償日止計算之法定遲延利息，為有理由，
    應予准許。又本件被告雖陳明願供擔保而聲請宣告免為假執
    行，惟因原告並未為假執行宣告之聲請，故本院認本件判決
    原告勝訴部分，原告既無從為假執行，則被告上開陳明願供
    擔保聲請宣告免為假執行部分，尚無必要，爰不另為准駁之
    裁判，附此敘明。
六、本件事證已臻明確，兩造所提其餘攻擊防禦方法及舉證，經
    審酌後核與本件判決結果不生影響，爰不一一論述，附此敘
    明。
七、訴訟費用負擔之依據：民事訴訟法第78條。
中    華    民    國  105     年    5     月    12    日
                  工程法庭  審判長法  官  林政佑
                                  法  官  吳坤芳
                                  法  官  蘇珈漪
以上正本係照原本作成。
如對本判決上訴，應於判決送達後20日內向本院提出上訴狀，若
委任律師提起上訴者，應一併繳納上訴審裁判費，否則本院得不
命補正逕行駁回上訴。
中    華    民    國   105    年    5     月    12    日
                                  書記官  簡吟倫
"""

content = f"""<content>{jud_content}</content>"""
rule = f"""
<rule>
1.原告(plaintiff)：說明在此案中被告為哪些公司或哪些人。
2.被告(defendant)：說明在此案中被告為哪些公司或哪些人。
3.調查對象(subject)：<subject>的內容，是我們的調查對象，通常是公司或個人。
4.摘要(summary)：簡述案情與判決結果，約200字。
4.調查對象的涉案狀況(case description)：須說明兩件事，其一說明調查對象在此篇判決書中是"原告"或"被告"或"不知道"，其二簡述調查對象的涉案狀況，約100字，若該公司與此篇判決書無關則輸出"與此案無關"
</rule>
"""
subject = f"<subject>{investigation_subject}</subject>"
prompt = f"""
<prompt>
你是銀行金融借貸的徵信調查人員正要調查<subject>提及的人物或公司，請你根據<content>中的裁判書內容，使用summarize_judgement工具生成JSON格式的回答。
</prompt>
"""

# https://docs.aws.amazon.com/bedrock/latest/userguide/model-parameters-anthropic-claude-messages.html
jud_tool = {
    "name": "summarize_judgement",
    "description": "Summarize judgement content.",
    "input_schema": {
        "type": "object",
        "properties": {
          "plaintiff": {
            "type": "string",
            "description": "Identifies the plaintiff(原告) in the judgment document."
          },
          "defendant": {
            "type": "string",
            "description": "Identifies the defendant(被告) in the judgment document."
          },
          "subject": {
              "type": "string",
              "description": "The subject of investigation in this judgment document.",
          },
          "subject_role": {
              "type": "string",
              "description": "The role of the subject of investigation in this judgment document.",
              "enum": ["原告", "被告", "關係人", "無關", "不知道"]
          },
          "chinese_summary": {
            "type": "string",
            "description": "Provides a summary of this judgment document."
          },
          "case_description": {
            "type": "string",
            "description": "Briefly describes the involvement of the subject of investigation in this judgment document."
          },
          "risk_score": {
            "type": "integer",
            "description": "Determines the risk of financial lending by the bank to the subject of investigation, ranging from 1 to 100.",
            "minimum": 1,
            "maximum": 100
          },          
          # "supporting_business_unit": {
          #     "type": "string",
          #     "description": "The internal business unit that this email should be routed to.",
          #     "enum": ["Sales", "Operations", "Customer Service", "Fund Management"]
          # },
          # "customer_names": {
          #     "type": "array",
          #     "description": "An array of customer names mentioned in the email.",
          #     "items": { "type": "string" }
          # },
          # "sentiment_towards_employees": {
          #     "type": "array",
          #     "items": {
          #         "type": "object",
          #         "properties": {
          #             "employee_name": {
          #                 "type": "string",
          #                 "description": "The employee's name."
          #             },
          #             "sentiment": {
          #                 "type": "string",
          #                 "description": "The sender's sentiment towards the employee.",
          #                 "enum": ["Positive", "Neutral", "Negative"]
          #             }
          #         }
          #     }
          # }
        },
        "required": [
            "plaintiff",
            "defendant",
            "subject",
            "subject_role",
            "chinese_summary",
            "case_description",
            "risk_score"
        ]
    }
}

# Claude 3 Haiku
body = {
    "anthropic_version": "bedrock-2023-05-31",
    "max_tokens": 4000,
    "system": rule,
    # "stop_sequences": [""],
    "messages": [
        # {
        #     "role": "assistant",
        #     "content": [
        #         {"type": "text", "text": rule}
        #     ]
        # },
        {
            "role": "user",
            "content": [
                {"type": "text", "text": content},
                {"type": "text", "text": subject},
                {"type": "text", "text": prompt},                
            ]
        }
    ],
    "temperature": 0.5,
    # "top_p": 0.999,
    # "top_k ": 500,
    "tools": [jud_tool],
    "tool_choice": {
        "type": "tool",
        "name": "summarize_judgement"
    }
}
kwargs = {
    "body": json.dumps(body),
    "contentType": "application/json",
    "accept": "application/json",
    "modelId": 'anthropic.claude-3-sonnet-20240229-v1:0',#"anthropic.claude-3-haiku-20240307-v1:0", #'anthropic.claude-3-5-sonnet-20240620-v1:0'
    # "trace": 'DISABLED', #'ENABLED'|'DISABLED',
    # "guardrailIdentifier":None, #'string',
    # "guardrailVersion":None, #'string'
}
```
